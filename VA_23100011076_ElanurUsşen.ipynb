```python
# =========================================================
# PROJE AÇIKLAMASI
# =========================================================
#Bu Python kodu, kentsel yeşil alanlara erişim ve yaşam tarzı değişkenlerini 
#kullanarak bireylerin fiziksel aktivite düzeyini tahmin 
#etmeyi ve bu değişkenlerin stres ve mutluluk üzerindeki etkilerini analiz
#etmeyi amaçlamaktadır.
#
# Projenin temel amacı, anket verilerinden yararlanarak bireylerin
# ruh sağlığı durumunu tahmin edebilecek sınıflandırma modelleri
# geliştirmek, bu modelleri karşılaştırmak ve en başarılı modeli
# belirlemektir.
#
# Kod genel olarak şu mantıkla çalışmaktadır:
#
# 1. Gerekli Python kütüphaneleri içe aktarılır.
#    numpy ve pandas veri işleme için, matplotlib ve seaborn
#    görselleştirme için, scipy istatistiksel analizler için,
#    scikit-learn ise makine öğrenmesi işlemleri için kullanılır.
#    missingno kütüphanesi ise eksik verilerin görsel olarak
#    incelenmesini sağlar.
#
# 2. Veri seti CSV dosyasından okunur.
#    Kod, veri dosyasını birkaç farklı olası dosya yolunda arar.
#    Dosya bulunursa pandas ile DataFrame yapısına aktarılır.
#    Eğer dosya bulunamazsa kullanıcıya açıklayıcı bir hata mesajı
#    verilir.
#
# 3. Veri seti hakkında ilk inceleme yapılır.
#    İlk 5 gözlem görüntülenir, veri setinin satır-sütun boyutu,
#    değişken türleri ve temel istatistiksel özetleri incelenir.
#    Bu adım, veri hakkında genel fikir edinmek için kullanılır.
#
# 4. Eksik veri analizi yapılır.
#    Hangi sütunlarda kaç adet eksik değer olduğu hesaplanır.
#    Eksik veriler hem heatmap hem de missingno grafikleriyle
#    görselleştirilir. Böylece eksik değerlerin rastgele mi yoksa
#    belirli bir yapıya göre mi oluştuğu yorumlanabilir.
#
# 5. Sayısal ve kategorik değişkenler ayrılır.
#    Sayısal değişkenler matematiksel işlemlerde ve ölçeklendirmede,
#    kategorik değişkenler ise encoding işlemlerinde kullanılır.
#    Hedef değişken olarak ruh sağlığı durumunu temsil eden
#    Mental_Health_Condition sütunu seçilir. Eğer bu sütun yoksa
#    veri setindeki son sütun hedef değişken kabul edilir.
#
# 6. Aykırı değer analizi yapılır.
#    Sayısal değişkenlerde IQR yöntemi kullanılarak alt ve üst sınırlar
#    hesaplanır. Bu sınırların dışında kalan gözlemler aykırı değer
#    olarak değerlendirilir. Ayrıca boxplot grafikleri ile değişkenlerin
#    aykırı değer yapısı görsel olarak incelenir.
#
# 7. Normal dağılım analizi yapılır.
#    Sayısal değişkenlerin normal dağılıma uygun olup olmadığını
#    incelemek için Shapiro-Wilk testi uygulanır. p-değeri 0.05'ten
#    büyükse değişken normal dağılıma yakın kabul edilir; küçükse
#    normal dağılmadığı yorumu yapılır.
#
# 8. Hedef değişkenin sınıf dağılımı incelenir.
#    Ruh sağlığı durumu sınıflarının veri setinde dengeli mi yoksa
#    dengesiz mi dağıldığı countplot grafiği ve value_counts çıktısı
#    ile analiz edilir. Bu adım sınıflandırma başarısını yorumlamak
#    açısından önemlidir.
#
# 9. Korelasyon analizi yapılır.
#    Sayısal değişkenler arasındaki doğrusal ilişkiler korelasyon
#    matrisi ile hesaplanır ve heatmap grafiğiyle gösterilir.
#    Bu sayede değişkenler arasında güçlü pozitif veya negatif
#    ilişkiler olup olmadığı incelenir.
#
# 10. Modelleme için veri hazırlanır.
#     Veri seti bağımsız değişkenler (X) ve hedef değişken (y) olarak
#     ayrılır. 
#     Hedef değişken olan Mental_Health_Condition başlangıçta
#     Low, Moderate ve High olmak üzere üç sınıftan oluşmaktadır.
#     Modelleme aşamasında bu değişken ikili sınıflandırmaya dönüştürülmüştür:
#     Low ve Moderate sınıfları 0, High sınıfı ise 1 olarak kodlanmıştır.
#
# 11. Ön işleme pipeline yapısı kurulur.
#     Sayısal değişkenler için eksik veriler IterativeImputer ile
#     doldurulur, aykırı değerlerin etkisini azaltmak için Winsorizer
#     uygulanır ve veriler StandardScaler ile ölçeklendirilir.
#     Kategorik değişkenlerde eksik değerler en sık görülen değerle
#     doldurulur ve OneHotEncoder ile sayısal forma dönüştürülür.
#
# 12. Veri sızıntısını önlemek için tüm ön işleme adımları Pipeline
#     ve ColumnTransformer yapıları içine alınır. Böylece eğitim
#     verisinde öğrenilen dönüşümler test verisine aynı kurallarla
#     uygulanır. Bu yaklaşım model değerlendirmesinin daha güvenilir
#     olmasını sağlar.
#
# 13. Özellik seçimi yapılır.
#     SelectKBest ve mutual_info_classif yöntemi kullanılarak hedef
#     değişkenle en ilişkili özellikler seçilir. Bu işlem gereksiz
#     değişkenlerin etkisini azaltır ve modelin daha verimli çalışmasına
#     yardımcı olur.
#
# 14. PCA ile boyut indirgeme uygulanır.
#     PCA, çok sayıda değişkeni daha az sayıda bileşenle temsil etmeye
#     yarar. Kodda açıklanan varyans oranı %90 olacak şekilde bileşenler
#     seçilir. Böylece verideki bilginin büyük kısmı korunurken boyut
#     azaltılır.
#
# 15. Üç farklı sınıflandırma modeli kurulur:
#     - KNN (K-Nearest Neighbors)
#     - Random Forest
#     - Gaussian Naive Bayes
#
#     Bu modeller aynı ön işleme, özellik seçimi ve PCA adımlarından
#     geçirilerek karşılaştırılır.
#
# 16. GridSearchCV ile hiperparametre optimizasyonu yapılır.
#     Her model için farklı parametre kombinasyonları denenir.
#     RepeatedStratifiedKFold çapraz doğrulama yöntemi kullanılarak
#     modellerin daha güvenilir şekilde değerlendirilmesi sağlanır.
#     Başarı ölçütü olarak weighted F1-score kullanılır.
#
# 17. En başarılı model seçilir.
#     GridSearchCV sonucunda en yüksek ortalama weighted F1 skoruna
#     sahip model final model olarak belirlenir.
#
# 18. PCA sonuçları görselleştirilir.
#     İlk iki ana bileşen üzerinden eğitim verisinin dağılımı çizilir.
#     Ayrıca PCA bileşenlerinin açıkladığı varyans oranları grafikle
#     gösterilir.
#
# 19. Modeller istatistiksel olarak karşılaştırılır.
#     Friedman testi kullanılarak modellerin çapraz doğrulama skorları
#     arasında istatistiksel olarak anlamlı fark olup olmadığı incelenir.
#
# 20. Final model test verisi üzerinde değerlendirilir.
#     Accuracy, precision, recall ve F1-score gibi başarı ölçütleri
#     hesaplanır. Ayrıca classification report ile her sınıf için
#     ayrıntılı performans sonuçları elde edilir.
#
# 21. Hata matrisi oluşturulur.
#     Modelin hangi sınıfları doğru, hangi sınıfları yanlış tahmin ettiği
#     confusion matrix ile görselleştirilir. Hem normal hem normalize
#     edilmiş hata matrisi çizilir.
#
# 22. # Hedef değişken modelleme öncesinde ikili sınıfa dönüştürüldüğü için
# ROC eğrisi ve AUC değeri hesaplanır.
#
# 23. Özellik seçimi sonuçları raporlanır.
#     Model tarafından seçilen en önemli değişkenler ve bu değişkenlerin
#     skorları listelenir. Böylece hedef değişken üzerinde hangi
#     özelliklerin daha etkili olduğu yorumlanabilir.
#
# 24. Sonuçta tüm modellerin performansları özet tablo halinde sunulur.
#     Bu tablo modellerin çapraz doğrulama başarısını, eğitim başarısını
#     ve test başarısını karşılaştırmayı sağlar.
#
# Genel olarak bu kod, veri analizi sürecini baştan sona uygulayan
# bütüncül bir çalışmadır. Önce veri tanınır ve temizlenir, sonra
# keşifsel veri analizi yapılır, ardından makine öğrenmesi modelleri
# oluşturulur, optimize edilir, karşılaştırılır ve final modelin
# performansı ayrıntılı olarak değerlendirilir.
# =========================================================

# =========================================================
# URBAN GREEN SPACE AND MENTAL HEALTH
# VERI ANALIZI DERSI PROJESI - TAMAMLANMIS SURUM
# =========================================================

# ---------------------------------------------------------
# 0. GEREKLI KUTUPHANELER
# ---------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")

import os
import sys
import subprocess
import importlib.util

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import shapiro, friedmanchisquare, wilcoxon
from scipy.stats.mstats import winsorize

from sklearn.base import BaseEstimator, TransformerMixin, clone
from sklearn.model_selection import (
    train_test_split,
    GridSearchCV,
    RepeatedStratifiedKFold,
    cross_val_score
)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.decomposition import PCA

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_curve,
    auc,
)
from sklearn.impute import KNNImputer

# missingno yoksa kurmayı dene
def ensure_package(package_name):
    if importlib.util.find_spec(package_name) is None:
        print(f"{package_name} paketi bulunamadi, kuruluyor...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package_name])

try:
    import missingno as msno
except:
    ensure_package("missingno")
    import missingno as msno

# Grafik ayarlari
plt.rcParams["figure.figsize"] = (8, 5)
sns.set_style("whitegrid")
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

# ---------------------------------------------------------
# 1. OZEL DONUSTURUCU: WINSORIZE (DÜZENLENMİŞ HALİ)
# ---------------------------------------------------------
class Winsorizer(BaseEstimator, TransformerMixin):
    def __init__(self, lower_limit=0.05, upper_limit=0.05):
        self.lower_limit = lower_limit
        self.upper_limit = upper_limit
        self.lower_bounds_ = None
        self.upper_bounds_ = None
        self.feature_names_in_ = None 

    def fit(self, X, y=None):
        X = pd.DataFrame(X).copy()
        self.feature_names_in_ = X.columns.tolist() 
        self.lower_bounds_ = X.quantile(self.lower_limit)
        self.upper_bounds_ = X.quantile(1 - self.upper_limit)
        return self

    def transform(self, X):
        X = pd.DataFrame(X).copy()
        return X.clip(lower=self.lower_bounds_, upper=self.upper_bounds_, axis=1).values

    # HATA ALMANIZI ENGELLEYEN KRİTİK METOT:
    def get_feature_names_out(self, input_features=None):
        return np.array(self.feature_names_in_)

# ---------------------------------------------------------
# 2. VERIYI YUKLEME
# ---------------------------------------------------------
possible_paths = [
    r"archive (2)\urban_green_space_and_mental_health_survey_dataset.csv",
    "urban_green_space_and_mental_health_survey_dataset.csv",
    "./urban_green_space_and_mental_health_survey_dataset.csv"
]

df = None
for path in possible_paths:
    if os.path.exists(path):
        df = pd.read_csv(path)
        print(f"Veri dosyasi yuklendi: {path}")
        break

if df is None:
    raise FileNotFoundError(
        "CSV dosyasi bulunamadi. Lutfen veri seti dosyasini notebook ile ayni klasore koy "
        "ya da dogru dosya yolunu possible_paths listesine ekle."
    )

# Sütun isimlerini temizle
df.columns = df.columns.str.strip()

# Hedef degiskeni otomatik tespit et
target_col = "Mental_Health_Condition" if "Mental_Health_Condition" in df.columns else df.columns[-1]

print("\n==================== 1. ILK BAKIS ====================")
print("Veri seti ilk 5 gozlem:")
print(df.head())

print("\nBoyut:", df.shape)
print("\nVeri tipleri:")
print(df.info())

print("\nTemel istatistikler:")
print(df.describe(include="all").T)

# ---------------------------------------------------------
# 3. EDA - EKSIK VERI ANALIZI
# ---------------------------------------------------------
print("\n==================== 2. EKSİK VERİ ANALİZİ ====================")
print("Eksik veri sayilari (islem oncesi):")
print(df.isnull().sum())

missing_summary = pd.DataFrame({
    "Eksik_Sayi": df.isnull().sum(),
    "Eksik_Yuzde": (df.isnull().sum() / len(df)) * 100
}).sort_values(by="Eksik_Yuzde", ascending=False)

print("\nEksik veri ozet tablosu:")
print(missing_summary)

plt.figure(figsize=(12, 5))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap="viridis")
plt.title("Eksik Veri Isi Haritasi - Islem Oncesi")
plt.show()

print("\nMissingno Matrix:")
msno.matrix(df)
plt.show()

print("\nMissingno Bar:")
msno.bar(df)
plt.show()

print("\nMissingno Heatmap:")
msno.heatmap(df)
plt.show()

# Sayisal ve kategorik sutunlar
# Önce modelden atılacak sütunları kesinleştiriyoruz (8. bölümde kullandığın liste ile aynı olmalı)
excluded_cols = ["respondent_id", "city", "occupation", "income_bracket", "gender", target_col]
eda_X = df.drop(columns=excluded_cols, errors="ignore")

numeric_cols = eda_X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = eda_X.select_dtypes(exclude=[np.number]).columns.tolist()

# ---------------------------------------------------------
# 4. AYKIRI DEGER ANALIZI (EDA)
# ---------------------------------------------------------
print("\n==================== 3. AYKIRI DEGER ANALİZİ ====================")

def iqr_outliers(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = len(dataframe[(dataframe[column] < lower) | (dataframe[column] > upper)])
    return lower, upper, outlier_count

if len(numeric_cols) > 0:
    print("\nIQR Analiz Sonuclari:")
    for col in numeric_cols:
        alt, ust, sayi = iqr_outliers(df, col)
        print(f"{col}: Alt Sinir={alt:.3f}, Ust Sinir={ust:.3f}, Aykiri Sayisi={sayi}")

    for col in numeric_cols:
        plt.figure(figsize=(8, 3.5))
        sns.boxplot(x=df[col])
        plt.title(f"{col} - Boxplot (Islem Oncesi)")
        plt.show()

# Ders Notu Referansı: Gorsellestirme_hafta6.ipynb (Violin Plot kullanımı)
if len(numeric_cols) > 0:
    plt.figure(figsize=(12, 6))
    sns.violinplot(x=target_col, y=numeric_cols[0], data=df, palette="muted")
    plt.title(f"{target_col} ve {numeric_cols[0]} Arasındaki Yoğunluk Dağılımı (Violin Plot)")
    plt.xticks(rotation=45)
    plt.show()

# ---------------------------------------------------------
# 5. NORMALLIK ANALIZI (EDA)
# ---------------------------------------------------------
print("\n==================== 4. NORMALLİK ANALİZİ ====================")
normality_results = []

for col in numeric_cols:
    sample_data = df[col].dropna()
    if len(sample_data) > 5000:
        sample_data = sample_data.sample(5000, random_state=42)

    if len(sample_data) >= 3:
        stat, p_value = shapiro(sample_data)
        decision = "Normal" if p_value > 0.05 else "Normal Degil"
        normality_results.append([col, stat, p_value, decision])

normality_df = pd.DataFrame(
    normality_results,
    columns=["Degisken", "Shapiro_Stat", "p_degeri", "Yorum"]
)
print(normality_df)

# Ornek donusum gorseli (ilk sayisal sutun varsa)
if len(numeric_cols) > 0:
    col = numeric_cols[0]
    temp_series = df[col].dropna().copy()
    shifted = temp_series - temp_series.min() + 1
    log_transformed = np.log1p(shifted)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(temp_series, kde=True, ax=axes[0])
    axes[0].set_title(f"Orijinal Dagilim - {col}")
    sns.histplot(log_transformed, kde=True, ax=axes[1])
    axes[1].set_title(f"Log Donusumu Sonrasi - {col}")
    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------
# 6. HEDEF DEGISKEN DAGILIMI
# ---------------------------------------------------------
print("\n==================== 5. HEDEF DEGISKEN DAGILIMI ====================")
plt.figure(figsize=(8, 5))
sns.countplot(x=target_col, data=df, palette="viridis")
plt.title(f"{target_col} Sinif Dagilimi")
plt.xticks(rotation=45)
plt.show()

print("\nSinif dagilimi:")
print(df[target_col].value_counts(dropna=False))

# ---------------------------------------------------------
# 7. KORELASYON ANALIZI (SADECE SAYISAL DEGİSKENLER)
# ---------------------------------------------------------
print("\n==================== 6. KORELASYON ANALİZİ ====================")
# Mevcut 7. bölümdeki heatmap kısmını şununla değiştirin:
if len(numeric_cols) > 1:
    plt.figure(figsize=(12, 8))
    corr_matrix = df[numeric_cols].corr()
    # center=0 ve coolwarm paleti ders notlarında vurgulanmıştı
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, linewidths=.5)
    plt.title("Sayısal Değişkenler Arası Ayrışan (Diverging) Korelasyon Analizi")
    plt.show()

# ---------------------------------------------------------
# 8. MODELLING ICIN VERI HAZIRLIGI
# ---------------------------------------------------------
print("\n==================== 7. MODELLEME İÇİN HAZIRLIK ====================")

# X ve y
# 'city' gibi çok fazla kategori içeren sütunlar 500 satırlık veride ezbere neden olur
drop_cols = ["city", "occupation", "income_bracket", "gender"]
# Bu sütunlar modelleme dışında bırakılmıştır.
# Özellikle city gibi yüksek kategorili değişkenler küçük veri setlerinde
# modelin ezber yapmasına ve genellenebilirliğin düşmesine neden olabilir.
print("Hedef degisken:", target_col)

# Önce hedef değişkeni dönüştür
df[target_col] = df[target_col].map({
    "Low": 0,
    "Moderate": 0,
    "High": 1
})

y = df[target_col]

# SONRA X oluştur
X = df.drop(columns=[target_col, "respondent_id"] + drop_cols, errors="ignore").copy()

numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\nX_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)

# ---------------------------------------------------------
# 9. ONISLEME PIPELINE'LARI
# ---------------------------------------------------------
# Sayisal pipeline
numeric_transformer = Pipeline(steps=[
    ("imputer", IterativeImputer(random_state=42)),
    ("winsorizer", Winsorizer(lower_limit=0.05, upper_limit=0.05)),
    ("scaler", StandardScaler())
])

# Kategorik pipeline
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ],
    remainder="drop"
)

# ---------------------------------------------------------
# 10. OZELLIK SECIMI + PCA + MODELLER
# ---------------------------------------------------------
print("\n==================== 8. MODEL VE HİPERPARAMETRE OPTİMİZASYONU ====================")

# Not:
# - Özellik seçimi: SelectKBest(mutual_info_classif)
# - Boyut indirgeme: PCA
# Böylece proje yönergesindeki "özellik seçimi" maddesi de karşılanmış olur.

common_steps = [
    ("preprocessor", preprocessor),
    ("feature_selection", SelectKBest(score_func=mutual_info_classif, k=8)),
    ("pca", PCA(n_components=0.90)) # Bilginin %90'ını koruyan en az bileşeni al
]
pipelines = {
    "KNN": Pipeline(steps=common_steps + [
        ("model", KNeighborsClassifier())
    ]),
    "RandomForest": Pipeline(steps=common_steps + [
        ("model", RandomForestClassifier(
            random_state=42,
            class_weight="balanced"
        ))
    ]),
    "GaussianNB": Pipeline(steps=common_steps + [
        ("model", GaussianNB())
    ])
}

param_grids = {
    "KNN": {
        "model__n_neighbors": [35, 45, 55], # Komşu sayısını artırdık (küçük k ezberletir)
        "model__weights": ["uniform"],      # 'distance' yerine 'uniform' kullanarak gürültüyü azalttık
        "model__metric": ["euclidean", "manhattan"]
    },
    "RandomForest": {
        "model__n_estimators": [100, 150],
        "model__max_depth": [3, 4],          # Derinliği 5'ten 3-4'e düşürdük (Budama)
        "model__min_samples_leaf": [15, 20], # Bir yaprakta en az 15-20 örnek olmalı
        "model__max_features": ["sqrt"],
        "model__bootstrap": [True]
    },
    "GaussianNB": {
        "model__var_smoothing": np.logspace(-1, -5, num=5) # Smoothing değerini artırdık
    }
}

cv_strategy = RepeatedStratifiedKFold(n_splits=5, n_repeats=2, random_state=42)

best_models = {}
grid_results = []

for model_name in pipelines.keys():
    print(f"\n>>> {model_name} icin GridSearchCV basladi...")

    grid = GridSearchCV(
        estimator=pipelines[model_name],
        param_grid=param_grids[model_name],
        scoring="f1_weighted",
        cv=cv_strategy,
        n_jobs=-1,
        return_train_score=True
    )

    grid.fit(X_train, y_train)

    best_models[model_name] = grid.best_estimator_
    grid_results.append({
        "Model": model_name,
        "Best_CV_F1_Weighted": grid.best_score_,
        "Best_Params": grid.best_params_
    })

    print(f"En iyi skor: {grid.best_score_:.4f}")
    print(f"En iyi parametreler: {grid.best_params_}")

results_df = pd.DataFrame(grid_results).sort_values(by="Best_CV_F1_Weighted", ascending=False)
print("\nGridSearchCV Sonuc Tablosu:")
print(results_df)

best_model_name = results_df.iloc[0]["Model"]
best_model = best_models[best_model_name]

print(f"\nSecilen final model: {best_model_name}")

# ---------------------------------------------------------
# 11. PCA GORSELLESTIRME (EN IYI PIPELINE'DAN)
# ---------------------------------------------------------
print("\n==================== 9. PCA GÖRSELLEŞTİRME ====================")
# PCA görselleştirmesi yalnızca eğitim verisi üzerinde yapılmıştır.
# Bu bölüm model performansı hesaplamak için değil,
# verinin iki ana bileşen üzerindeki dağılımını göstermek için kullanılır.
if "feature_selection" in best_model.named_steps and "pca" in best_model.named_steps:
    preprocessed_train = best_model.named_steps["preprocessor"].fit_transform(X_train, y_train)

    if hasattr(preprocessed_train, "toarray"):
        preprocessed_train = preprocessed_train.toarray()

    selected_train = best_model.named_steps["feature_selection"].fit_transform(preprocessed_train, y_train)
    pca_train = best_model.named_steps["pca"].fit_transform(selected_train)

    if pca_train.shape[1] >= 2:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(
            x=pca_train[:, 0],
            y=pca_train[:, 1],
            hue=y_train,
            palette="Set1",
            alpha=0.7
        )
        
        plt.title("PCA ile İlk İki Bileşen Üzerinde Veri Dağılımı")
        plt.xlabel("PC1")
        plt.ylabel("PC2")
        plt.show()
        # PCA varyans grafiği
        pca_model = best_model.named_steps["pca"]

        plt.figure(figsize=(8, 5))
        plt.plot(np.cumsum(pca_model.explained_variance_ratio_), marker="o")
        plt.xlabel("Bileşen Sayısı")
        plt.ylabel("Kümülatif Açıklanan Varyans")
        plt.title("PCA Kümülatif Açıklanan Varyans")
        plt.grid(True)
        plt.show()
else:
    print("PCA ve feature selection kaldırıldığı için bu bölüm atlandı.")

from itertools import combinations

print("\n==================== 10. 30 SEED İLE İSTATİSTİKSEL MODEL KIYASLAMASI ====================")

# 0-10000 arası 30 farklı, karışık seed üret
np.random.seed(42)
all_seeds = np.arange(0, 10001)
np.random.shuffle(all_seeds)
seed_list = all_seeds[:30]

print("Kullanılan seed değerleri:")
print(seed_list)

seed_results = []

for seed in seed_list:
    X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
        X, y,
        test_size=0.20,
        random_state=int(seed),
        stratify=y
    )

    for model_name, model in best_models.items():
        temp_model = clone(model)
        temp_model.fit(X_train_s, y_train_s)

        y_pred_s = temp_model.predict(X_test_s)

        seed_results.append({
            "Seed": int(seed),
            "Model": model_name,
            "Accuracy": accuracy_score(y_test_s, y_pred_s),
            "Precision_Weighted": precision_score(y_test_s, y_pred_s, average="weighted", zero_division=0),
            "Recall_Weighted": recall_score(y_test_s, y_pred_s, average="weighted", zero_division=0),
            "F1_Weighted": f1_score(y_test_s, y_pred_s, average="weighted", zero_division=0)
        })

seed_results_df = pd.DataFrame(seed_results)

print("\n30 Seed Sonuç Tablosu:")
print(seed_results_df)

summary_seed_df = seed_results_df.groupby("Model").agg(
    F1_Mean=("F1_Weighted", "mean"),
    F1_Std=("F1_Weighted", "std"),
    F1_Median=("F1_Weighted", "median"),
    Accuracy_Mean=("Accuracy", "mean")
).sort_values(by="F1_Mean", ascending=False)

print("\nModel Özet Tablosu:")
print(summary_seed_df)

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=seed_results_df, x="Model", y="F1_Weighted")
sns.stripplot(data=seed_results_df, x="Model", y="F1_Weighted", color="black", alpha=0.5)
plt.title("30 Farklı Seed Üzerinde Model Performansları")
plt.ylabel("Weighted F1 Score")
plt.xlabel("Model")
plt.show()

# Friedman testi için wide format
friedman_df = seed_results_df.pivot(index="Seed", columns="Model", values="F1_Weighted")

friedman_stat, friedman_p = friedmanchisquare(
    *[friedman_df[col] for col in friedman_df.columns]
)

print("\nFriedman Testi:")
print(f"Test istatistiği: {friedman_stat:.4f}")
print(f"p-değeri: {friedman_p:.6f}")

alpha = 0.05

if friedman_p < alpha:
    print("H0 reddedilir: Algoritmalar arasında istatistiksel olarak anlamlı fark vardır.")

    print("\nİkili Wilcoxon Karşılaştırmaları:")
    pairwise_results = []

    model_names = list(friedman_df.columns)

    for m1, m2 in combinations(model_names, 2):
        stat, p_value = wilcoxon(friedman_df[m1], friedman_df[m2])

        mean_m1 = friedman_df[m1].mean()
        mean_m2 = friedman_df[m2].mean()

        better_model = m1 if mean_m1 > mean_m2 else m2

        pairwise_results.append({
            "Model_1": m1,
            "Model_2": m2,
            "Mean_Model_1": mean_m1,
            "Mean_Model_2": mean_m2,
            "Wilcoxon_Stat": stat,
            "p_value": p_value,
            "Significant": "Evet" if p_value < alpha else "Hayır",
            "Better_Model": better_model if p_value < alpha else "Anlamlı fark yok"
        })

    pairwise_df = pd.DataFrame(pairwise_results)
    print(pairwise_df)

else:
    print("H0 reddedilemez: Algoritmalar arasında istatistiksel olarak anlamlı fark yoktur.")
    print("Bu durumda ikili karşılaştırma yapmak zorunlu değildir.")

# ---------------------------------------------------------
# 13. FINAL MODEL DEGERLENDIRME - TRAIN VE TEST
# ---------------------------------------------------------
print("\n==================== 11. FİNAL MODEL DEĞERLENDİRME ====================")

best_model.fit(X_train, y_train)
X_train_processed = best_model.named_steps["preprocessor"].transform(X_train)
print("\nEksik veri kontrolü (imputation sonrası):")
print(pd.DataFrame(X_train_processed).isnull().sum())
print("Toplam eksik değer:", pd.DataFrame(X_train_processed).isnull().sum().sum())
# Tahminler
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

# Basari olcutleri fonksiyonu
def compute_metrics(y_true, y_pred, average_type="weighted"):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average=average_type, zero_division=0),
        "Recall": recall_score(y_true, y_pred, average=average_type, zero_division=0),
        "F1_Score": f1_score(y_true, y_pred, average=average_type, zero_division=0)
    }

train_metrics = compute_metrics(y_train, y_train_pred)
test_metrics = compute_metrics(y_test, y_test_pred)

metrics_table = pd.DataFrame([train_metrics, test_metrics], index=["Train", "Test"])
print("\nTrain/Test Basari Tablosu:")
print(metrics_table)

print("\nTrain Classification Report:")
print(classification_report(
    y_train,
    y_train_pred,
    target_names=["Low/Moderate", "High"],
    zero_division=0
))

print("\nTest Classification Report:")
print(classification_report(
    y_test,
    y_test_pred,
    target_names=["Low/Moderate", "High"],
    zero_division=0
))

# ---------------------------------------------------------
# 14. HATA MATRISI
# ---------------------------------------------------------
print("\n==================== 12. HATA MATRİSİ ====================")

fig, ax = plt.subplots(figsize=(7, 5))
ConfusionMatrixDisplay.from_estimator(
    best_model,
    X_test,
    y_test,
    display_labels=["Low/Moderate", "High"],
    cmap="Blues",
    ax=ax
)
plt.title(f"Hata Matrisi - {best_model_name} (Test)")
plt.xticks(rotation=45)
plt.show()

fig, ax = plt.subplots(figsize=(7, 5))
ConfusionMatrixDisplay.from_estimator(
    best_model,
    X_test,
    y_test,
    display_labels=["Low/Moderate", "High"],
    cmap="Greens",
    normalize="true",
    ax=ax
)
plt.title(f"Normalize Hata Matrisi - {best_model_name} (Test)")
plt.xticks(rotation=45)
plt.show()

# ---------------------------------------------------------
# 15. ROC / AUC (SADECE IKILI SINIFLANDIRMA ISE)
# ---------------------------------------------------------
print("\n==================== 13. ROC / AUC ANALİZİ ====================")

is_binary = len(np.unique(y)) == 2

if is_binary:
    plt.figure(figsize=(8, 6))

    auc_results = []

    for model_name, model in best_models.items():
        

        # Olasılık tahmini al
        if hasattr(model, "predict_proba"):
            y_prob = model.predict_proba(X_test)[:, 1]
        elif hasattr(model, "decision_function"):
            y_prob = model.decision_function(X_test)
        else:
            continue

        # ROC hesapla
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)

        # ROC çiz
        plt.plot(fpr, tpr, label=f"{model_name} (AUC={roc_auc:.3f})")

        # AUC kaydet
        auc_results.append({
            "Model": model_name,
            "AUC": roc_auc
        })

    # Random çizgi
    plt.plot([0, 1], [0, 1], "k--")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Eğrileri Karşılaştırması")
    plt.legend()
    plt.show()

    # 🔹 AUC BAR GRAFİĞİ
    auc_df = pd.DataFrame(auc_results)

    plt.figure(figsize=(8, 5))
    sns.barplot(data=auc_df, x="Model", y="AUC")
    plt.title("Modellerin AUC Skorları")
    plt.ylim(0, 1)
    plt.show()

    # 🔹 En iyi model AUC
    best_row = auc_df.sort_values(by="AUC", ascending=False).iloc[0]
    print(f"En iyi AUC: {best_row['Model']} -> {best_row['AUC']:.4f}")

else:
    print("Veri seti çok sınıflı olduğu için ROC/AUC atlandı.")

# ---------------------------------------------------------
# 16. OZELLIK SECIMI SONUCLARI
# ---------------------------------------------------------
print("\n==================== 14. ÖZELLİK / DEĞİŞKEN BİLGİSİ ====================")

preprocessor_fitted = clone(preprocessor)
preprocessor_fitted.fit(X_train, y_train)

feature_names = preprocessor_fitted.get_feature_names_out()

print("Modelde kullanılan toplam özellik sayısı:", len(feature_names))

feature_info_df = pd.DataFrame({
    "Feature": feature_names
})

print("\nModelde kullanılan özellikler:")
print(feature_info_df)

# Eğer modelde feature_selection varsa ayrıca seçilenleri göster
if "feature_selection" in best_model.named_steps:
    X_train_preprocessed = preprocessor_fitted.transform(X_train)

    if hasattr(X_train_preprocessed, "toarray"):
        X_train_preprocessed = X_train_preprocessed.toarray()

    selector = clone(best_model.named_steps["feature_selection"])
    selector.fit(X_train_preprocessed, y_train)

    support_mask = selector.get_support()

    selected_features = pd.DataFrame({
        "Selected_Feature": feature_names[support_mask]
    })

    print("\nSeçilen özellikler:")
    print(selected_features)

    if hasattr(selector, "scores_") and selector.scores_ is not None:
        feature_scores = pd.DataFrame({
            "Feature": feature_names,
            "Score": selector.scores_
        }).sort_values(by="Score", ascending=False)

        print("\nİlk 20 özellik skoru:")
        print(feature_scores.head(20))

        plt.figure(figsize=(10, 6))
        top_features = feature_scores.head(15).sort_values(by="Score")
        plt.barh(top_features["Feature"], top_features["Score"])
        plt.title("En Yüksek Skorlu 15 Özellik")
        plt.xlabel("Mutual Information Score")
        plt.show()

else:
    print("\nFeature selection kaldırıldığı için tüm ön işlenmiş özellikler modele verilmiştir.")

# ---------------------------------------------------------
# 17. SONUC OZET TABLOSU
# ---------------------------------------------------------
print("\n==================== 15. GENEL SONUÇ TABLOSU ====================")


summary_rows = []
for model_name, model in best_models.items():
    # Modeli tekrar eğit (en iyi parametrelerle)
    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    
    # Skorları hesapla
    train_acc = accuracy_score(y_train, train_pred)
    test_acc = accuracy_score(y_test, test_pred)

    summary_rows.append({
        "Model": model_name,
        "CV_F1_Weighted_Mean": results_df.loc[
         results_df["Model"] == model_name,
         "Best_CV_F1_Weighted"
        ].values[0],
        "Train_Accuracy": train_acc,
        "Test_Accuracy": test_acc,
        # BURAYA EKLEDİK:
        "Train_Test_Gap": train_acc - test_acc, 
        "Test_F1_Weighted": f1_score(y_test, test_pred, average="weighted", zero_division=0)
    })
# DataFrame oluştur
summary_df = pd.DataFrame(summary_rows)

# Sadece sıralama yap (Hata aldığın kısım burasıydı)
summary_df = summary_df.sort_values(
    by="Train_Test_Gap", # İstersen en az farka göre sıralatabilirsin
    ascending=True
)
plt.figure(figsize=(9, 5))
sns.barplot(data=summary_df, x="Model", y="Train_Test_Gap")
plt.title("Overfitting Analizi (Train - Test)")
plt.show()
print(summary_df)
plt.figure(figsize=(9, 5))
sns.barplot(data=summary_df, x="Model", y="Test_F1_Weighted")
plt.title("Modellerin Test F1 Skorları")
plt.ylim(0, 1)
plt.show()

# ---------------------------------------------------------
# 18. RAPORDA KULLANABILECEGIN OTOMATIK YORUM CIKTILARI
# ---------------------------------------------------------
print("\n==================== 16. RAPOR İÇİN KISA YORUM METİNLERİ ====================")

print("1) Eksik veri analizi tamamlandi; sayisal degiskenlerde IterativeImputer, kategorik degiskenlerde en sik gozlenen deger ile atama stratejisi kullanildi.")
print("2) Aykiri deger etkisini azaltmak icin winsorization uygulandi.")
print("3) Veri sizintisini onlemek amaciyla modelleme on islemleri pipeline icerisine alindi.")
print("4) Ozellik secimi icin SelectKBest(mutual_info_classif), boyut indirgeme icin PCA kullanildi.")
print("5) KNN, Random Forest ve Gaussian Naive Bayes algoritmalari GridSearchCV ile optimize edildi.")
print("6) Modeller RepeatedStratifiedKFold ile degerlendirildi.")
print("7) Modeller arasindaki farkin anlamliligi Friedman testi ile incelendi.")
print("8) Final model icin hem egitim hem test basari metrikleri raporlandi.")

print("\nTum analizler basariyla tamamlandi.")
```
